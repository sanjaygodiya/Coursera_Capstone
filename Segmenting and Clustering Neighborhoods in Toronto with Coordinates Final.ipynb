```python
import pandas as pd
import requests 
```

Importing web page from url

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
web_txt=requests.get(url).text
```

Parsing HTML page

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(web_txt,'xml')
table = soup.find('table',{'class':'wikitable sortable'})
```

Converting table data into array 

```python
table_rows=table.find_all('tr')
data = []
for row in table_rows:
    td=[]
    for t in row.find_all('td'):
        td.append(t.text.strip())
    data.append(td)
```

```python
df=pd.DataFrame(data,columns=['PostalCode', 'Borough', 'Neighborhood'])
```

```python
df.shape
```

Deleting rows with 'Not assigned' value in Borough column

```python
df.drop(df[df['Borough']=='Not assigned'].index,inplace=True)
df.drop(df[df.index==0].index,inplace=True)
df=df.reset_index()
df.drop(columns=['index'],inplace=True)
```

Creating a function to explore 100 venues at each loaction within 500 radius

```python
LIMIT=100
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID,CLIENT_SECRET,VERSION,lat,lng,radius,LIMIT)
            
        # GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # append only relevant information for each nearby venue
        venues_list.append([(name,lat,lng,v['venue']['name'],v['venue']['location']['lat'],v['venue']['location']['lng'],
                             v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Calling above funtion to create dataframe of venues

```python
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],latitudes=toronto_data['Latitude'],longitudes=toronto_data['Longitude'])
toronto_venues.head()
```

Analyzing each neighborhood with one shot encoding

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="")
toronto_onehot['Neighborhood1'] = toronto_venues['Neighborhood'] 
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
toronto_onehot.columns=toronto_onehot.columns.str.replace('_','')
toronto_onehot.head()
```

Let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood1').mean().reset_index()
toronto_grouped.head()
```

Finding top 5 venues for each neighbourhood

```python
for neigh in toronto_grouped['Neighborhood1']:
    print("----"+neigh+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood1'] == neigh].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(5))
    print('\n')
```

Creating dataframe for above result

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False) 
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
#Creating dataframe for top 10 venues
num_top_venues = 10
indicators = ['st', 'nd', 'rd']
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood1']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```
